```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('import_export.csv')
data.set_index(data.columns[0], inplace=True)
display(data)
```

```python

# Convert missing values ("-") to NaN and change to numeric
data = data.astype(float)

# Plot heatmap
plt.figure(figsize=(18, 6))
sns.heatmap(data, cmap="coolwarm", annot=True, fmt=".2f", linewidths=0.5)

# Labels & Title
plt.title("LNG Import & Export (Billion Cubic Feet)")
plt.xlabel("Importer")
plt.ylabel("Exporter")

plt.show()
```
