```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('total-primary-energy-supply-by-fuel-1971-and-2019.csv')
data.set_index('Year', inplace=True)
display(data)

ax = data.plot(kind='bar', stacked=True, figsize=(10, 6))
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='center', fontsize=10, color='black')
plt.ylabel('%')
plt.xlabel('Year')
plt.title('Primary Energy Supply by Fuel Type (1971 - 230 EJ vs 2019 - 606 EJ)')
plt.legend(title="Fuel", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```
