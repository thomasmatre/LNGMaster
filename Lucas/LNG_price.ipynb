```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "datatest.csv"
df = pd.read_csv(file_path)

# Clean and convert the date column
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y', errors='coerce')

# Rename columns for easier access
df.rename(columns={
    'Global price of Natural Gas, Asia (start month, MMBTU)': 'Asia',
    'Global price of Natural Gas, EU (start month, MMBTU)': 'EU',
    'Global price of Natural Gas, US Henry Hub (start month, MMBTU)': 'US'
}, inplace=True)

# Plotting the LNG prices
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Asia'], label='Asia', linewidth=2)
plt.plot(df['Date'], df['EU'], label='Europe', linewidth=2)
plt.plot(df['Date'], df['US'], label='US (Henry Hub)', linewidth=2)
plt.title('Global LNG Prices by Region')
plt.xlabel('Date')
plt.ylabel('Price ($/MMBTU)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Fix column names for easier access
df.rename(columns={
    'Spread\nEU-US': 'Spread EU-US',
    'Spread\nASIA-US': 'Spread ASIA-US',
    'Spread\nASIA-EU': 'Spread ASIA-EU'
}, inplace=True)

# Plot the spread values
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Spread ASIA-US'], label='Asia - US', color='orange')
plt.plot(df['Date'], df['Spread EU-US'], label='EU - US', color='blue')
plt.plot(df['Date'], df['Spread ASIA-EU'], label='Asia - EU', color='green')

plt.title('LNG Price Spreads Between Regions')
plt.xlabel('Date')
plt.ylabel('Price Spread (USD/MMBtu)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reload the correct dataset
df = pd.read_csv("datatest.csv")

# Convert 'Date' column to datetime, assuming format is like "Jan-15"
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y', errors='coerce')
df = df.dropna(subset=['Date'])

# Sort and set index
df = df.sort_values('Date')
df.set_index('Date', inplace=True)

# Plot LNG Carrier Newbuilding Prices
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['LNG Carrier Newbuilding Prices (end month, $m)'], marker='o', linestyle='-')
plt.title('LNG Carrier Newbuilding Prices (20152025)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (Million $)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "datatest.csv"
df = pd.read_csv(file_path)

# Clean column names (remove extra whitespace)
df.columns = df.columns.str.strip()

# Clean and convert the date column
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y', errors='coerce')

# Plot 1: LNG Orderbook (CBM)
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['LNG Orderbook (start month, CBM)'], color='steelblue', linewidth=2)
plt.title('LNG Carrier Orderbook Over Time (CBM)')
plt.xlabel('Date')
plt.ylabel('Orderbook (CBM)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Approximate Ships on Order
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Aprox.\nShip on order'], color='darkorange', linewidth=2)
plt.title('Approximate Number of LNG Carriers on Order Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Ships')
plt.grid(True)
plt.tight_layout()
plt.show()
```
